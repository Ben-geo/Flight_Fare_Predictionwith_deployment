```python
# Departure time is when a plane leaves the gate. 
# Similar to Date_of_Journey we can extract values from Dep_Time

# Extracting Hours
data["Dep_hour"] = pd.to_datetime(data["Dep_Time"]).dt.hour

# Extracting Minutes
data["Dep_min"] = pd.to_datetime(data["Dep_Time"]).dt.minute

# Now we can drop Dep_Time as it is of no use
data.drop(["Dep_Time"], axis = 1, inplace = True)
```

```python
data["Dep_day"] = pd.to_datetime(data["Date_of_Journey"]).dt.day
data["Dep_month"] = pd.to_datetime(data["Date_of_Journey"]).dt.month
data["Dep_year"] = pd.to_datetime(data["Date_of_Journey"]).dt.year

# Now we can drop Dep_Time as it is of no use
data.drop(["Date_of_Journey"], axis = 1, inplace = True)
```

```python
data["arrival_hour"] = pd.to_datetime(data["Arrival_Time"]).dt.day
data["arrival_minute"] = pd.to_datetime(data["Arrival_Time"]).dt.month


data.drop(["Arrival_Time"], axis = 1, inplace = True)
```
